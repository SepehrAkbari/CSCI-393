```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(7, 7)
qc.h(range(7))
qc.measure(range(7), range(7))

qc.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
transpiled_qc = transpile(qc, backend)

num_shots = 1000

job = backend.run(transpiled_qc, shots=num_shots)
```

```python
result = job.result()
counts = result.get_counts()
```

```python
random_integers = []
for outcome, count in counts.items():
    for _ in range(count):
        random_integers.append(int(outcome, 2) + 1)

random_integers = random_integers[:1000]

print(random_integers)
plot_histogram(counts)
```

```python
import pandas as pd

df = pd.DataFrame(random_integers, columns=["Random Integers"])

file_path = "Q_randoms.csv"
df.to_csv(file_path, index=False)

print(f"Random integers saved to {file_path}")
```
